```python
class Money:

  def __init__(self, rubles = 0: int, kopec = 0: int):

    self.__rubles = rubles
    self.__kopec = kopec

  @property
  def rubles(self):

    return self.__rubles

  @property
  def kopec(self):

    return self.__kopec

  @rubles.setter
  def rubles(self, amount: int):

    if 0 <= amount:
      self.__rubles = amount
    else:
      print('Недопустимое количество рублей')

  @kopec.setter
  def kopec(self, amount: int):

    if 0 <= amount < 100:
      self.__rubles = amount
    else:
      print('Недопустимое количество копеек')

  def print_amount(self):

    if self.__kopec < 10:
      print(self.rubles, ',', 0, self.kopec, sep = '')
    else:
      print(self.rubles, ',', self.kopec, sep = '')

  def add_money(self, money: object):

    total_self = self.__rubles * 100 + self.__kopec
    total_money = money.__rubles * 100 + money.__kopec
    new_amount = total_self + total_money
    self.__rubles = new_amount // 100
    self.__kopec = new_amount % 100

  def reduce_money(self, money: object):

    total_self = self.__rubles * 100 + self.__kopec
    total_money = money.__rubles * 100 + money.__kopec
    if total_self >= total_money:
      new_amount = total_self - total_money
      self.__rubles = new_amount // 100
      self.__kopec = new_amount % 100
    else:
      print('нельзя отнять столько, сколько нет))')

  def divide_money(self, n: int):

    total = self.__rubles * 100 + self.__kopec
    temp = Money()
    new_amount = int(total / n)
    temp.__rubles = int(new_amount // 100)
    temp.__kopec = int(new_amount % 100)
    return temp

  def divide_by_money(self, money: object):

    total_self = self.__rubles * 100 + self.__kopec
    total_money = money.__rubles * 100 + money.__kopec
    amount = total_self // total_money
    return amount

  def multip_money(self, n: int):

    total = self.__rubles * 100 + self.__kopec
    temp = Money()
    new_amount = int(total * n)
    temp.__rubles = int(new_amount // 100)
    temp.__kopec = int(new_amount % 100)
    return temp

  def compare_money(self, money: object):

    total_self = self.__rubles * 100 + self.__kopec
    total_money = money.__rubles * 100 + money.__kopec
    if total_self == total_money:
      print(f'{total_self / 100} = {total_money / 100}')
    elif total_self < total_money:
      print(f'{total_self / 100} < {total_money / 100}')
    else:
      print(f'{total_self / 100} > {total_money / 100}')
```

```python
wallet = Money()
wallet.rubles = 400
wallet.print_amount()

wallet1 = Money(50, 60)
wallet1.print_amount()

wallet.add_money(wallet1)
wallet.print_amount()

wallet.add_money(wallet1)
wallet.print_amount()

wallet.reduce_money(wallet1)
wallet.print_amount()

wallet2 = wallet.divide_money(5)
wallet2.print_amount()

wallet.divide_by_money(wallet2)

wallet2 = wallet2.multip_money(5)
wallet2.print_amount()

wallet1.compare_money(wallet2)
```

```python
class Matrix:

  def __init__(self, matrix: list):

    self.__matrix = matrix
    self.__rows = len(matrix)
    self.__colums = len(matrix[0])

  def show_matrix(self):

    for row in self.__matrix:
      for item in row:
        print(item, end = ' ')
      print()

  def add(self, matr: object):

    if self.__rows == matr.__rows and self.__colums == matr.__colums:
      for i in range(self.__rows):
        for j in range(self.__colums):
          self.__matrix[i][j] += matr.__matrix[i][j]
    else:
      print('матрицы не одинакового размера')

  def multip(self, n):

    for i in range(self.__rows):
        for j in range(self.__colums):
          self.__matrix[i][j] *= n

```

```python
matr1 = Matrix([[1, 2, 3], [4, 5, 6]])
matr1.show_matrix()

print()
matr1.add(matr1)
matr1.show_matrix()

print()
matr1.multip(2)
matr1.show_matrix()
```

```python
class Vector:

  def __init__(self, lst = []: list):

    self.__vect = lst

  def show(self):

    print(self.__vect)

  def add(self, vect: object):

    if len(self.__vect) == len(vect.__vect):
      for i in range(len(self.__vect)):
        self.__vect[i] += vect.__vect[i]

  def multip(self, n: int):

    for i in range(len(self.__vect)):
        self.__vect[i] *= n

  def subtract(self, vect: object):

    if len(self.__vect) == len(vect.__vect):
      for i in range(len(self.__vect)):
        self.__vect[i] -= vect.__vect[i]

  def scalar_multip(self, vect: object) -> int:

    res = 0
    if len(self.__vect) == len(vect.__vect):
      for i in range(len(self.__vect)):
        temp = self.__vect[i] * vect.__vect[i]
        res += temp
    return res
```

```python
some_vector = Vector([1, 2, 3, 4, 5])
another_vector = Vector([2, 4, 6 , 8, 10])
some_vector.show()

some_vector.add(another_vector)
some_vector.show()

some_vector.subtract(another_vector)
some_vector.show()

some_vector.multip(3)
some_vector.show()

some_vector.scalar_multip(another_vector)
```
